```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_2000 = pd.read_csv("Data/final_tmdb_data_2000.csv.gz")
df_2000
```

```python
df_2001 = pd.read_csv("Data/final_tmdb_data_2001.csv.gz")
df_2001
```

```python
# Concatenate both the DataFrames
df_concat = pd.concat([df_2000, df_2001], axis = 0)
df_concat
```

```python
budget_filter = df_concat["budget"] >= 1.0
revenue_filter = df_concat["revenue"] >= 1.0
df_concat = df_concat[budget_filter & revenue_filter]
```

```python
cert_list = ["R", "PG-13", "PG", "G"]
cert_df = df_concat[df_concat.certification.isin(cert_list)]
cert_df
```

```python
x = ["Movies with Financial Info", "Movies with Certification Categories"]
y = [len(df_concat), len(cert_df)]
fig, ax = plt.subplots()

ax.bar(x, y, edgecolor = "k")

ax.set_ylabel("Movies")
ax.set_title("Movie Count")
```

```python
g_cert = cert_df["certification"] == "G"
g_cert_df = cert_df[g_cert]
g_revenue = g_cert_df["revenue"].mean()
g_budget = g_cert_df["budget"].mean()
```

```python
r_cert = cert_df["certification"] == "R"
r_cert_df = cert_df[r_cert]
r_revenue = r_cert_df["revenue"].mean()
r_budget = r_cert_df["budget"].mean()
```

```python
pg13_cert = cert_df["certification"] == "PG-13"
pg13_cert_df = cert_df[pg13_cert]
pg13_revenue = pg13_cert_df["revenue"].mean()
pg13_budget = pg13_cert_df["budget"].mean()
```

```python
pg_cert = cert_df["certification"] == "PG"
pg_cert_df = cert_df[pg_cert]
pg_revenue = pg_cert_df["revenue"].mean()
pg_budget = pg_cert_df["budget"].mean()
```

```python
revenue_avg = [r_revenue, pg13_revenue, pg_revenue, g_revenue]
fig, ax = plt.subplots()

ax.bar(cert_list, revenue_avg, edgecolor = "k")

ax.set_ylabel("Revenue in $")
ax.set_xlabel("Certifications")
ax.set_title("Certification Revenue Averages")
```

```python
budget_avg = [r_budget, pg13_budget, pg_budget, g_budget]
fig, ax = plt.subplots()

ax.bar(cert_list, budget_avg, edgecolor = "k")

ax.set_ylabel("Budget in $")
ax.set_xlabel("Certifications")
ax.set_title("Certification Budget Averages")
```
