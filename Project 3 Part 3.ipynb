```python
import pandas as pd
import pymysql
pymysql.install_as_MySQLdb()
from sqlalchemy import create_engine
from sqlalchemy_utils import create_database, database_exists
from sqlalchemy.types import *
```

```python
basics = pd.read_csv("Data/title_basics.csv.gz", low_memory = False)
basics.head()
```

```python
ratings = pd.read_csv("Data/title_ratings.csv.gz", low_memory = False)
ratings.head()
```

```python
api_results2000 = pd.read_csv("Data/final_tmdb_data_2000.csv.gz", low_memory = False)
api_results2000.head()
```

```python
api_results2001 = pd.read_csv("Data/final_tmdb_data_2001.csv.gz", low_memory = False)
api_results2001.head()
```

```python
## create a col with a list of genres
basics['genres_split'] = basics['genres'].str.split(',')
basics
```

```python
exploded_genres = basics.explode('genres_split')
exploded_genres
```

```python
unique_genres = basics["genres_split"].explode().unique()
unique_genres
```

```python
unique_genres = sorted(exploded_genres['genres_split'].unique())
unique_genres
```

```python
# Create a new title_genres table
title_genres = exploded_genres[["tconst", "genres_split"]].copy()
title_genres.head()
```

```python
## Making the genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
title_genres['genre_id'] = title_genres['genres_split'].map(genre_map)
title_genres = title_genres.drop(columns = 'genres_split')
```

```python
genre = pd.DataFrame({"Genre_Name": genre_map.keys(), "Genre_ID": genre_map.values()})
genre.head()
```

```python
# Discard unecessary information

basics.drop(columns = ["originalTitle", "isAdult", "titleType", "genres"])
```

```python
## Calculate max string lengths for object columns
key_len = basics['tconst'].fillna('').map(len).max()
title_len = basics['primaryTitle'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_schema = {
    "tconst": String(key_len+1), 
    "primaryTitle": Text(title_len+1),
    'startYear':Float(),
    'endYear':Float(),
    'runtimeMinutes':Integer()}
```

```python
basics.to_sql("title_basics", engine, dtype = df_schema, if_exists = "replace", index = False)
ratings.to_sql("title_ratings", engine, dtype = df_schema, if_exists = "replace", index = False)
api_results2000.to_sql("tmbd2000", engine,dtype = df_schema,  if_exists = "replace", index = False)
api_results2001.to_sql("tmbd2001", engine,dtype = df_schema,  if_exists = "replace", index = False)
genres.to_sql("genres", engine, dtype = df_schema,  if_exists = "replace", index = False)
title_genres.to_sql("title_genres", engine , dtype = df_schema, if_exists = "replace", index = False)
```
