```python
api_results2001 = pd.read_csv("Data/final_tmdb_data_2001.csv.gz", low_memory = False)
api_results2001.head()
```

```python
tmdb_data_combined = pd.concat([api_results2000, api_results2001])
tmdb_data_combined.head()
```

```python
tmdb_data_combined = tmdb_data_combined.loc[tmdb_data_combined['imdb_id']!='0']
```

```python
## create a col with a list of genres
basics['genres_split'] = basics['genres'].str.split(',')
basics
```

```python
exploded_genres = basics.explode('genres_split')
exploded_genres
```

```python
unique_genres = basics["genres_split"].explode().unique()
unique_genres
```

```python
unique_genres = sorted(exploded_genres['genres_split'].unique())
unique_genres
```

```python
# Create a new title_genres table
title_genres = exploded_genres[["tconst", "genres_split"]].copy()
title_genres.head()
```

```python
## Making the genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
title_genres['genre_id'] = title_genres['genres_split'].map(genre_map)
title_genres = title_genres.drop(columns = 'genres_split')
```

```python
genre_lookup = pd.DataFrame({"Genre_Name": genre_map.keys(), "Genre_ID": genre_map.values()})
genre_lookup.head()
```

```python
# Discard unecessary information

basics = basics.drop(columns = ["originalTitle", "isAdult", "titleType", "genres", "genres_split"])
basics.info()
```

```python
tmdb_data_combined["imdb_id"].duplicated().sum()
tmdb_data_combined = tmdb_data_combined.drop_duplicates(subset = "imdb_id")
```

```python
tmdb_df = tmdb_data_combined[["imdb_id", "budget", "revenue", "certification"]]
tmdb_df.head()
```

```python
from sqlalchemy.types import *
## Calculate max string lengths for object columns
key_len = basics['tconst'].fillna('').map(len).max()
title_len = basics['primaryTitle'].fillna('').map(len).max()
genre_name_len = genre_lookup['Genre_Name'].fillna('').map(len).max()
certification_len = tmdb_df['certification'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_schema = {
    'tconst': String(key_len+1), 'primaryTitle': Text(title_len+1),
    'start_year': Float(), 'runtime_minutes': Integer(),
    'average_rating': Float(), 'num_votes': Integer(),
    'genre_id': Integer(), 'Genre_Name': Text(genre_name_len+1),
    'imdb_id': String(key_len+1), 'revenue': Float(),
    'budget': Float(), 'certification': Text(certification_len+1)}
```

```python
basics.to_sql("title_basics", engine, dtype = df_schema, if_exists = "replace", index = False)
ratings.to_sql("title_ratings", engine, dtype = df_schema, if_exists = "replace", index = False)
tmdb_df.to_sql("tmdb_data", engine,dtype = df_schema,  if_exists = "replace", index = False)
genre_lookup.to_sql("genres", engine, dtype = df_schema,  if_exists = "replace", index = False)
title_genres.to_sql("title_genres", engine , dtype = df_schema, if_exists = "replace", index = False)
```

```python
engine.execute('ALTER TABLE genres ADD PRIMARY KEY (`Genre_ID`);')
engine.execute('ALTER TABLE title_basics ADD PRIMARY KEY (`tconst`);')
engine.execute('ALTER TABLE title_ratings ADD PRIMARY KEY (`tconst`);')
engine.execute('ALTER TABLE tmdb_data ADD PRIMARY KEY (`imdb_id`);')
```
